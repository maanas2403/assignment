```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import requests
%matplotlib inline
```

```python
def get_json_data(api_link):
    response = requests.get(api_link)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print("Failed to retrieve data from the API")
        return None
```
